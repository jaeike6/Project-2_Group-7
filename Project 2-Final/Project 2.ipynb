```python
import pandas as pd
from sqlalchemy import create_engine


```

```python
salary_df=pd.read_csv("../resources/nba_salary_final.csv")
salary_df= salary_df[["player_name","salary"]]
salary_df
```

```python
stats_df=pd.read_csv("../resources/nba_stats_final.csv")
stats_df
```

```python
db_connection="postgres:postgres@localhost:5432/Project2_db"
engine=create_engine(f"postgresql://{db_connection}")
```

```python
stats_df.to_sql(name="nba_stats",con=engine,index=False,if_exists="append")
```

```python
salary_df.to_sql(name="nba_salary",con=engine,index=False,if_exists="append")
```

```python
Team_df=pd.read_csv("../resources/Team.csv")
Team_df
```

```python
Team_df.to_sql(name="team_stats",con=engine,index=False,if_exists="append")
```
