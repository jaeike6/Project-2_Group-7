```python
#import dependancies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#extract data from CSV files
csv_file = "Original files/team_records.csv"
team_data_df = pd.read_csv(csv_file)
team_data_df.head()

csv_file2 = "Original files/Team_Key.csv"
team_key_df = pd.read_csv(csv_file2)
```

```python
#clean NBA season record data to include the season where player 
#statistics and salary are available

season=['2017-18']
season_df=team_data_df[team_data_df["Season"].isin(season)]
season_df.head()
```

```python
#Check dataframe
teams=season_df['Team'].unique()

print(teams)
```

```python
#Clean dataframe
teams_1718=season_df[['Team', 'W/L%', 'Finish']].copy()
teams_1718.head(30)
```

```python
team_key_df.head(30)
```

```python
team_combined=pd.merge(teams_1718, team_key_df, on="Team")
team_combined.head()
```

```python
#combine dataframes to add Tm variable
team_final=team_combined[['Tm', 'W/L%', 'Finish']].copy()
team_final.head(30)
```

```python
#export final clean team dataset to be merged with the player statistics 
#data using TM
team_final.to_csv('Clean data/Team.csv', index=False)                 
```
